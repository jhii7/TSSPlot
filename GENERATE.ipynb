```python
%pylab inline
import seaborn as sns
import scipy.stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Generate dataframe
df = pd.read_csv("~/lab5/annotations/tss_histogram.txt", sep="\t")
df.columns = ["Distance from Center", "Oct4 Coverage", "Oct4 tags", "Oct4 -tags", "Sox2 Coverage", "Sox2 tags", "Sox2 -tags", "Klf4 Coverage", "Klf4 tags", "Klf4 -tags", "H3K4me2 Coverage", "H3K4me2 tags", "H3K4me2 -tags", "H3K27ac Coverage", "H3K27ac tags", "H3K27ac -tags"]
df
```

```python
df.plot('Distance from Center', y = ['Oct4 Coverage', 'Sox2 Coverage', 'Klf4 Coverage', 'H3K4me2 Coverage', 'H3K27ac Coverage'])
plt.ylabel("Relative read count")
plt.title("Composite Plot")
plt.savefig('CompositePlot.jpg')
plt.show()
```
